```python
from sage.graphs.graph_plot import GraphPlot

def numDistinctMultiplicities(mults):
    distinctMults = []
    for m in mults:
        if (m in distinctMults) == false:
            distinctMults.append(m)
    return len(distinctMults)

n = 16
trees = graphs.trees(n)
points = []
numMultsVsMaxMultPoints = []
counter = 0
for T in trees:
    mults = list(T.distances_distribution().values())
    numMults = numDistinctMultiplicities(mults)
    diam = T.diameter()
    if diam == +Infinity:
        diam = n
    points.append([diam,numMults])
    maxMult = max(mults)*n*(n-1)/2
    numMultsVsMaxMultPoints.append([numMults,maxMult])
#print(points)
LowBound = []
for d in range(1,n):
    LowBound.append([d,floor((d^2)/(n-1))])
plot = scatter_plot(points, alpha = 0.01, facecolor = 'black', frame = True, axes_labels = ['Diameter', 'Number of Distinct Multiplicities']) + scatter_plot(LowBound, markersize = 10, facecolor='red')
show(plot)
numMultsVsMaxMultPlot = scatter_plot(numMultsVsMaxMultPoints, alpha = 0.05, facecolor = 'black', frame = True, axes_labels = ['Number of Distinct Multiplicities','Max Multiplicity'], axes = False)
#+ text('Number of Distinct Multiplicities', (floor(n/2),-7), bounding_box={'boxstyle':'round', 'fc':'w'}) +  text('Max Multiplicity', (-1,(n-1)*(n-2)/4), bounding_box={'boxstyle':'round', 'fc':'w'}, rotation = "vertical")
show(numMultsVsMaxMultPlot)

UpperBoundExamples = []
for pair in numMultsVsMaxMultPoints:
    include = True
    for comparedPair in numMultsVsMaxMultPoints:
        if pair[0] == comparedPair[0]:
            if pair[1] < comparedPair[1]:
                include = False
    if include == True and (pair not in UpperBoundExamples) == True:
        UpperBoundExamples.append(pair)

print(UpperBoundExamples)
for T in graphs.trees(n):
    mults = list(T.distances_distribution().values())
    numMults = numDistinctMultiplicities(mults)
    maxMult = max(mults)*n*(n-1)/2
    for pair in UpperBoundExamples:
        if pair[0] == numMults and pair[1] == maxMult:
            show(T)
            print(pair)
    
```
