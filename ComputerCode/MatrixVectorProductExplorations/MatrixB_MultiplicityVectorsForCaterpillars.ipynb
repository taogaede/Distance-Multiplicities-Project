```python
from sage.plot.colors import red, blue, lime
MatrixPoints = []
for a1 in range(1, 15):
    for a2 in range(1,15):
        for a3 in range(1,15):
            for a4 in range(1,15):
                for a5 in range(1,15):
                    n = a1 + a2 + a3 + a4 + a5 + 2
                    if n == 17:
                        v = vector([a1,a2,a3,a4,a5])
                        B = matrix([[a2,a3,a4,a5,0], [a3,a4,a5,0,0], [a4,a5,0,0,0], [a5,0,0,0,0]])
                        M = (B*v)/(n*(n-1)/2)
                        MatrixPoints.append(M)
                        #print(M)
plotList = []
for p in MatrixPoints:
    point = point3d([[p[0],p[1],p[2]]], alpha = 0.1, size = 25*p[3], color='red', aspect_ratio = [1,1,1], frame = True)
    plotList.append(point)              
CatScatterPlot = point3d([[0,0,0]], alpha = 0, facecolor='red', frame = True)
for s in plotList:
    CatScatterPlot = CatScatterPlot + s
show(CatScatterPlot)
        
```

```python
TreePoints = []
for n in range(17,18):
    for T in graphs.trees(n):
        if T.diameter() == 6:
            degSeq = T.degree_sequence()
            r = 0
            for i in degSeq:
                if i > 1:
                    r = r+1
            if r >= 6:
                D = T.distances_distribution()
                M = vector([D[3],D[4],D[5],D[6]])
                TreePoints.append(M)
plotList = []
for p in TreePoints:
    point = point3d([[p[0],p[1],p[2]]], alpha = 0.1, size = 25*p[3], facecolor=(1,0,0), aspect_ratio = [1,1,1], frame = True)
    plotList.append(point)              
theScatterPlot = point3d([[0,0,0]], alpha = 0, facecolor='red', frame = True)
for s in plotList:
    theScatterPlot = theScatterPlot + s
show(theScatterPlot + CatScatterPlot)
```
